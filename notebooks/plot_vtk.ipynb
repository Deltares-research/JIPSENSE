```python
import pyvista as pv
```

```python
# 1. Read the VTK dataset
mesh = pv.read("/home/obandohe/JIPSENSE/JIPSENSE/data/output/OUTPUT_FILES_2026-06-25_17-23-06/mesh00000_vs.vtk")

# 2. Check available data arrays (scalars or vectors) inside your file
print("Available data arrays:", mesh.array_names)
```

```python


# 3. Initialize the Plotter object
plotter = pv.Plotter()

# 4. Add the mesh with customized parameters
# Replace 'your_scalar_name' with a string from mesh.array_names
plotter.add_mesh(
    mesh, 
    scalars=mesh.array_names[0],  # Color code by the first data array
    cmap="coolwarm",              # Choose a colormap
    show_edges=True,              # Outline the cell/mesh edges
    edge_color="black"
)

# 5. Add helpful visual anchors
plotter.add_axes()
plotter.add_scalar_bar(title="Data Scale")

# show xz view
plotter.view_xz()  # Set the view to XZ plane

# show grid
plotter.show_grid()


# 6. Render the window
plotter.show()
```
